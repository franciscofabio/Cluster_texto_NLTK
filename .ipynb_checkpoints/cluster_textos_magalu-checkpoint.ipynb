```python
#total de termos no documento
from nltk.stem import WordNetLemmatizer

def total_termos():

    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(w,'v') for w in text1]

    return len(set(lemmatized))

total_termos()
```

```python
#quais s√£o os termos mais frequentes 
def termos_frequentes():
  dist = nltk.FreqDist(text1)
  vocab1= dist.keys()
  freqwords = [w for w in vocab1 if len(w) > 3 and dist[w] >2]
  return sorted(freqwords)
termos_frequentes()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import seaborn as sns
import matplotlib as mtl
import matplotlib.pyplot as plt
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(comentarios["tweet"])
```

```python
X = normalize(X)
```

```python
# Cria a Curva de Cotovelo para encontrar o Numero Ideal de Clusters
from sklearn.cluster import KMeans
wcss = []

for i in range(2, 10):
    kmeans = KMeans(n_clusters = i, max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
# Mostra o Gr√°fico
plt.plot(range(2, 10), wcss)
plt.title('CThe Elbow Method')
plt.xlabel('Numero de Clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
plt.savefig('elbow.png', format='png')
```

```python
# vamos usar o n de cluster recomendado pelo metodo elbow n=6
kmeans = KMeans(
    n_clusters=6,
    n_init=15,
    max_iter=450,
    init = 'k-means++'
    
)
```

```python
#treinar 

kmeans.fit(X)
```

```python
comentarios['cluster'] = kmeans.labels_
comentarios
```

```python
comentarios['cluster'].value_counts(normalize=True)
```

```python
# Fun√ß√£o para limpar e tokenizar os clusters
def clear_text_cluster(vet):
    from string import punctuation
    
    text = str(vet.tolist())
    text = demoji.replace(text,"")
    text = text.replace('\\U0001f979',"")
    text = text.replace('\\u200d',"")
    tk = nltk.word_tokenize(text)
    tk = [w for w in tk if not w in set(stopwords.words('portuguese'))]
    tk = [w for w in tk if not w in punctuation]
    
    return tk

# Fun√ß√£o para mostrar os ngrams
def view_ngrams(token,n):
    r_gram = ngrams(token,n)
    r_top = []
    for grams in r_gram:
        r_top.append(list(grams))
    return r_top
```

```python
tk_cl0 = clear_text_cluster(cluster0['tweet'])
```
